```python
dummies_ = pd.get_dummies(streamHistory_df[['artistName', 'dayOfWeek', 'month', 'dayTime']])
```

```python
processedData = dummies_.join(streamHistory_df[['trackName', 'msPlayed']])
```

```python
processedData.head(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.linear_model import LogisticRegression
 

```

```python
X = processedData.drop('trackName', axis=1)
X.head()
```

```python
y = processedData['trackName']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
for i in range(1,11):
    dtree = DecisionTreeClassifier(criterion="entropy", max_depth = i).fit(X_train, y_train)
    pred_dtree = dtree.predict(X_test)
    print(accuracy_score(y_test, pred_dtree), '\t','max_depth =',i)
```

```python
for i in ['linear','poly','rbf','sigmoid']:
    svm_ = svm.SVC(kernel=i).fit(X_train,y_train)
    pred_svm = svm_.predict(X_test)
    print(accuracy_score(y_test, pred_svm), '\t',i)
```
